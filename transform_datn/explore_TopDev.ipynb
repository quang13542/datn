```python
time_unit = [
    'years',
    'year',
    'months',
    'month',
    'năm',
    'tháng'
]
def extract_year(year_exp):
    pattern = f"(\S+)\s({'|'.join(time_unit)})"
    matches = re.findall(pattern, year_exp)
    # print(pattern)
    res = []
    for match in matches:
        res.append(match[0] + ' ' + match[1])
    return res
```

```python
df['job_detail_year_of_exp'].map(lambda exp:
    [extract_year(year_exp) for year_exp in exp]
)
```
